```python
import datasets
from datasets import Dataset, DatasetDict
from datasets import load_from_disk 
from transformers import  AutoTokenizer , AutoModel
import numpy as np
from torch import nn
from tqdm.notebook import tqdm_notebook
from transformers import AdamW
import torch 
from sklearn.metrics import f1_score, classification_report
```

```python
model_checkpoint = "HooshvareLab/bert-base-parsbert-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
datasets = DatasetDict()
datasets["train"] =  load_from_disk("/content/data/dataset.train")
datasets["validation"] = load_from_disk("/content/data/dataset.dev")
datasets["test"] = load_from_disk("/content/data/dataset.test")
```

```python
import json

with open('/content/data/dataset.slots') as json_file:
    index_to_slot = json.load(json_file)
index_to_slot[str(len(index_to_slot))] = "pad"
slots_to_index = {v: k for k, v in index_to_slot.items()}

with open('/content/data/dataset.intents') as json_file:
    index_to_intent = json.load(json_file)
intent_to_index = {v: k for k, v in index_to_intent.items()}

num_intents = len(intent_to_index)
num_slots = len(slots_to_index)
```

preprocess dataset

```python
def preprocess(record):
  max_len = 512
  input_ids = np.zeros(512)
  attention_mask = np.zeros(512)
  slots =  np.ones(512) * int(slots_to_index["pad"])
  input_ids[0] = tokenizer.cls_token_id
  attention_mask[0] = 1
  intent = record["intent_num"]
  c = 1
  for idx , words in enumerate(record["utt"]):
    for word in tokenizer.encode(words,add_special_tokens=False):
      input_ids[c] = word
      attention_mask[c] = 1
      slots[c] = record["slots_num"][idx]
      c+=1
  input_ids[c] = tokenizer.sep_token_id
  return {"input_ids" : input_ids , "attention_mask" : attention_mask ,"intent" : intent , "slots" :slots }
```

```python
datasets["train"] = datasets["train"].map(preprocess ,remove_columns = datasets["train"].column_names)
datasets["validation"] = datasets["validation"].map(preprocess ,remove_columns = datasets["validation"].column_names)
datasets["test"] = datasets["test"].map(preprocess,remove_columns = datasets["test"].column_names)
```

#create joint model model

```python
class JointModel(nn.Module):
    
    def __init__(self ):
        super(JointModel, self).__init__()
        self.bert = AutoModel.from_pretrained(model_checkpoint)
        self.intent = nn.Linear(self.bert.config.hidden_size , num_intents)
        self.slot = nn.Linear(self.bert.config.hidden_size , num_slots)
        
    def forward(self , input_ids , attention_mask):
        output = self.bert(input_ids = input_ids  , attention_mask = attention_mask )
        slots = self.slot(output[0]) 
        intents = self.intent(output[1])
        return intents, slots
```

utils

```python
def get_non_pad(preds,reals):
  realsc = reals.cpu()
  non_pad_idx = np.logical_not(np.equal(realsc.cpu(),np.array(int(slots_to_index["pad"])))).nonzero().view(-1)
  return reals[non_pad_idx].tolist() , preds[non_pad_idx].tolist()
```

```python
def get_tp(pred_,real_):
  preds = pred_.argmax(axis=1).cpu().numpy()
  reals = real_.cpu().numpy()
  tp = 0
  for idx , pred in enumerate(preds):
    if pred == reals[idx] and pred != int(slots_to_index["pad"]):
      tp +=1
  num_words = len(reals)-np.sum(np.equal(reals,np.array(int(slots_to_index["pad"]))))
  return num_words , tp 
```

#train

```python
datasets["train"].set_format(type='torch')
datasets["validation"].set_format(type="torch")
datasets["test"].set_format(type="torch")
```

```python
def validation( model , val_loader , device = "cuda" , print_report = False ):
  val_loss = 0.0
  model.eval()
  num_words = 0
  tps = 0
  reals = []
  pads = []
  for batch in val_loader:
    input_ids = batch["input_ids"].to(device).to(torch.int64)
    attention_mask = batch["attention_mask"].to(device).to(torch.int64)
    real_intents = batch["intent"].to(device).view(-1).to(torch.int64)
    real_slots = batch["slots"].to(device).view(-1).to(torch.int64)
    pred_intent , pred_slot = model(input_ids = input_ids , attention_mask = attention_mask)
    pred_slot = pred_slot.view(-1, pred_slot.shape[-1])
    loss_intent = loss_fn(pred_intent , real_intents)
    loss_slots = loss_fn(pred_slot , real_slots)
    loss = (loss_intent + loss_slots) / 2
    val_loss += loss.item()
    numwords , tp = get_tp(pred_intent , real_intents)
    non_pad_reals , non_pad_preds = get_non_pad(pred_slot.argmax(axis=1).cpu().numpy(),real_slots)
    reals += non_pad_reals
    pads += non_pad_preds
    num_words += numwords
    tps += tp

  val_loss = val_loss / len(val_loader)
  accuracy = tps/num_words
  if print_report :
    reals = [index_to_slot[str(r)] for r in reals]
    pads = [index_to_slot[str(r)] for r in pads]
    print(classification_report(reals , pads , labels=list(slots_to_index.keys())))
  return val_loss ,accuracy
```

```python
def train(model , optimizer , loss_fn , train_loader , val_loader , test_loader , epochs , device = "cuda"):
  model.to(device)
  for epoch in range(epochs):
    training_loss = 0.0
    val_loss =0.0
    model.train()
    for batch in tqdm_notebook(train_loader):
      input_ids = batch["input_ids"].to(device).to(torch.int64)
      attention_mask = batch["attention_mask"].to(device).to(torch.int64)
      real_intents = batch["intent"].to(device).view(-1).to(torch.int64)
      real_slots = batch["slots"].to(device).view(-1).to(torch.int64)
      optimizer.zero_grad()
      pred_intent , pred_slot = model(input_ids = input_ids , attention_mask = attention_mask)
      pred_slot = pred_slot.view(-1, pred_slot.shape[-1])
      loss_intent = loss_fn(pred_intent , real_intents)
      loss_slots = loss_fn(pred_slot , real_slots)
      loss = (loss_intent + loss_slots) / 2
      loss.backward()
      optimizer.step()
      training_loss += loss.item()
    training_loss = training_loss / len(train_loader)

    
    val_loss , val_accuracy = validation(model , val_loader , device)

    print('Epoch: {},training loss: {:.2f} , validation loss: {:.2f} , validation intent detection accuracy: {:.2f}'.format(epoch , training_loss , val_loss,val_accuracy))
  test_loss , test_accuracy = validation(model , test_loader , device , print_report = True)
  print('test loss: {:.2f}  test intent detection accuracy: {:.2f}'.format( test_loss , test_accuracy))
```

#evaluate

```python
epoch = 10
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = JointModel()
optimizer = AdamW(model.parameters(), lr=2e-5) 
loss_fn = nn.CrossEntropyLoss(ignore_index=(int(slots_to_index["pad"])))

train_loader = torch.utils.data.DataLoader(datasets["train"] , batch_size=16)
val_loader = torch.utils.data.DataLoader(datasets["validation"] , batch_size=8)
test_loader = torch.utils.data.DataLoader(datasets["test"] , batch_size=8)

train(model , optimizer , loss_fn , train_loader , val_loader , test_loader , epoch , device )
```
